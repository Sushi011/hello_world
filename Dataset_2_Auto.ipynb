```python

# Assuming x_test, y_test, and model have been defined earlier
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
evs = explained_variance_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("Explained Variance Score:", evs)

# Add the specific lines for Linear Regression metrics
print("Linear Regression - Coefficient of determination (R²): {:.2f}%".format(r2 * 100))
print("Linear Regression - Mean Squared Error (MSE): {:.2f}".format(mse))
```
